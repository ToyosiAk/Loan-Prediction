**Independent Variable (Categorical)**

```python
plt.figure(1) 
plt.subplot(221)
train['Gender'].value_counts(normalize =True).plot.bar(title= 'Gender', figsize =(20, 10))

plt.subplot(222)
train['Married'].value_counts(normalize =True).plot.bar(title= 'Married', figsize =(20, 10))

plt.subplot(223)
train['Self_Employed'].value_counts(normalize =True).plot.bar(title= 'Self_Employed', figsize =(20, 10))

plt.subplot(224)
train['Credit_History'].value_counts(normalize =True).plot.bar(title= 'Credit_History', figsize =(20, 10))
plt.show()
```

It can be inferred that:
- 80% of the applicants in the dataset are male;
- Around 65% of the applicants in the dataset are married;
- Around 15% of the applicants in the dataset are self employed;
- Around 85% of applicants have repaid their debt;

**Independent Variables (Ordinal)**

```python
plt.figure(1) 
plt.subplot(131)
train['Dependents'].value_counts(normalize =True).plot.bar(title= 'Dependents', figsize =(20, 5))

plt.subplot(132)
train['Education'].value_counts(normalize =True).plot.bar(title= 'Education', figsize =(20, 5))

plt.subplot(133)
train['Property_Area'].value_counts(normalize =True).plot.bar(title= 'Property_Area', figsize =(20, 5))
plt.show()
```

It can be inferred that:

- Around 80% of the applicants in the dataset are graduates;
- Most of the applicants in the dataset are from Semiurban areas;

### Bivariate Analysis
**Categorical Independent Variable vs Target Variable**

```python
Gender = pd.crosstab(train['Gender'], train['Loan_Status'])
Gender.div(Gender.sum(1).astype(float), axis=0).plot(kind=  'bar', stacked= True, figsize= (4, 4))
```

```python
Married = pd.crosstab(train['Married'], train['Loan_Status'])
Married.div(Married.sum(1).astype(float), axis=0).plot(kind= 'bar', stacked= True, figsize= (4, 4))

Self_Employed = pd.crosstab(train['Self_Employed'], train['Loan_Status'])
Self_Employed.div(Self_Employed.sum(1).astype(float), axis=0).plot(kind= 'bar', stacked= True, figsize= (4, 4))

Credit_History = pd.crosstab(train['Credit_History'], train['Loan_Status'])
Credit_History.div(Credit_History.sum(1).astype(float), axis=0).plot(kind= 'bar', stacked= True, figsize= (4, 4))
plt.show()
```
